### merge

```python
train4.head()
```

```python
train4.shape
```

```python
# just merge on date first
train5 = train4.merge(hol_events, how='left', on=['date'])
train5
```

```python
t1_local = train4.merge(local, how='inner', left_on=['date','city'], right_on=['date','locale_name'])
t1_local
```

```python
t1_regional = train4.merge(regional, how='inner', left_on=['date','state'], right_on=['date','locale_name'])
t1_regional
```

```python
t1_national = train4.merge(national, how='inner', on='date')
t1_national
```

```python
# filter out dates that are in holiday_events
dates = hol_events['date'].unique()
```

```python
# to get row id that have dates in holiday but does not match in city/state to locale_name
id_holidays = train5[train5['date'].isin(dates)]['id'].tolist()
```

```python
t_no_hols = train5[~train5['date'].isin(dates)]
t_no_hols
```

```python
test = pd.concat([t_no_hols, t1_local, t1_regional, t1_national], axis=0).reset_index().drop(columns=['index']).sort_values(by='id')
test
```

```python
matched_ids = test['id'].unique().tolist()
```

```python
# add in rows that match holiday dates but not match location

not_matched_df = train4[(train4['id'].isin(id_holidays)) & (~train4['id'].isin(matched_ids))]
#train5[(train5['id'].isin(id_holidays)) & ((train5['city']) | ())]
```

```python
not_matched_df
```

```python
not_matched_df2 = not_matched_df.merge(hol_events, how='left', left_on=['date','city'], right_on=['date','locale_name'])
```

```python
not_matched_df2
```

```python
not_matched_df2['locale'].unique()
```

```python
# combine all
pd.concat([test,not_matched_df2], axis=0).reset_index().drop(columns=['index']).sort_values(by='id')
```
