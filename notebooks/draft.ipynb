```python
def fill_in_nulls(df):
    # check for NA
    df['transactions'].fillna(0, inplace=True)
    df['dcoilwtico'].fillna(0, inplace=True)
    df['transferred'].fillna(False, inplace=True)
    df.fillna('None', inplace=True)  # fillna for the rest are strings
    return df
```
